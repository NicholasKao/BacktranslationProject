```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = None
```

```python
df = pd.read_excel('Results.xls', sheet_name = 'Results')
```

```python
df['SpanishEditDist%'] = pd.to_numeric(df['SpanishEditDist%'])#.str.replace('%',''))
df['ChineseEditDist%'] = pd.to_numeric(df['ChineseEditDist%'])#.str.replace('%',''))
df['AfrikaansEditDist%'] = pd.to_numeric(df['AfrikaansEditDist%'])#.str.replace('%',''))
```

```python
fig, ax = plt.subplots(figsize = (12,8))
sns.lineplot(data = df, x = 'WordCount', y = 'SpanishEditDist%', marker = 'o', markersize = 10,
             linewidth = 2, color = 'b', label = 'Spanish', ax = ax)
sns.lineplot(data = df, x = 'WordCount', y = 'ChineseEditDist%', marker = 'o', markersize = 10,
             linewidth = 2, color = 'r', label = 'Chinese', ax = ax)
sns.lineplot(data = df, x = 'WordCount', y = 'AfrikaansEditDist%', marker = 'o', markersize = 10,
             linewidth = 2, color = 'g', label = 'Afrikaans', ax = ax)
ax.set_xlabel('Number of Words in Phrase', size = 12)
ax.set_ylabel('Edit Distance as % of Phrase Length', size = 12)
plt.title('Comparison of Edit Distance Percentage\n vs. Phrase Word Count Across Languages',         
         size = 15)
```
